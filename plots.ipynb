```python
import matplotlib.pyplot as plt
```

```python
import pickle as pkl
with open ('DKD_stat/ResNet8_Resnet18_tckd_alpha_0.5.pkl', 'rb') as f:
    train_losses, val_losses, train_accuracies, val_accuracies = pkl.load(f)

with open ('DKD_stat/ResNet8_Resnet18_tckd_alpha_0.25.pkl', 'rb') as f:
    train_losses1, val_losses1, train_accuracies1, val_accuracies1 = pkl.load(f)

with open ('DKD_stat/ResNet8_Resnet18_tckd_alpha_1.pkl', 'rb') as f:
    train_losses2, val_losses2, train_accuracies2, val_accuracies2 = pkl.load(f)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.plot(val_accuracies1, label='alpha=0.25')
plt.plot(val_accuracies, label='alpha=0.5')
plt.legend()
plt.title('Val Accuracies of ResNet8 Student model TCKD only')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
# plt.plot(train_losses, label='train')
plt.plot(val_losses, label='alpha=0.5')
plt.plot(val_losses1, label='alpha=0.25')
plt.plot(val_losses2, label='alpha=1.0')
plt.legend()
plt.title('Val Losses of ResNet8 Student model NCKD only')
plt.xlabel('Epoch')
plt.ylabel('Losses')
```
